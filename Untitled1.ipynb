```python
import sqlite3
import os
```

```python
os.getcwd()
```

```python
conn = sqlite3.connect(':memory:')
```

```python
c = conn.cursor()
```

```python
c.execute("""create table employees(first text, second text, pay integer )""")
```

```python
conn.commit()
```

```python
c.execute("select * from employees")
c.fetchall()
```

```python
# it can be accessed using contect manager.So it does not require explicitly to call commit()
# fetchall or fetchone or fetchmany keeps on fetching rows until more rows are left

with conn:
    c.execute("""insert into employees values(:first,:last,:pay)""",{'first':'puneet','last':'bansal','pay':10000})
c.execute("select * from employees")
c.fetchall()  
```
